```python
import torch
from diffusers import DDPMPipeline
import os
from tqdm import tqdm  # For progress bar

# Create output directory
output_dir = "generated_faces"
os.makedirs(output_dir, exist_ok=True)

# Load model
pipeline = DDPMPipeline.from_pretrained("otausendschoen/ddpm-our-faces-reduced")

# Move to GPU and use half precision
device = "cuda" if torch.cuda.is_available() else "cpu"
pipeline = pipeline.to(device)
if device == "cuda":
    pipeline = pipeline.to(torch.float16)

# Generate 100 images with different seeds
num_images = 100

# Use tqdm for progress tracking
for i in tqdm(range(num_images)):
    # Use a different seed for each image
    generator = torch.Generator(device=device).manual_seed(i)

    # Generate the image
    image = pipeline(generator=generator).images[0]

    # Save the image
    image_path = os.path.join(output_dir, f"face_{i:03d}.png")
    image.save(image_path)

print(f"Successfully generated {num_images} images in the '{output_dir}' directory")
```
